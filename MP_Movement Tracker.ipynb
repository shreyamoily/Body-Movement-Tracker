```python
import tensorflow as tf
import tensorflow_hub as hub
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
model = hub.load('https://tfhub.dev/google/movenet/multipose/lightning/1')
movenet = model.signatures['serving_default']
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
  ret, frame = cap.read() #reading the current frame where the video is at. Frame is the image that we want to capture

  cv2.imshow('Movenet Multipose', frame) #shows the captured image

  #exiting the frame, if 'q' is pressed the frame will be exited
  if cv2.waitKey(10) & 0xFF==ord('q'):
    break
cap.release()
cv2.destroyAllWindows()
```

```python
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```
